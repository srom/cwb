```python
import os
from pathlib import Path

import numpy as np
import pandas as pd
from rdkit import Chem

cwd = os.getcwd()
if cwd.endswith('notebook/PG'):
    os.chdir('../..')
    cwd = os.getcwd()

from src.pg_modelling.ligand_utils import sanitize_ligand_name
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('colorblind')
sns.set_style('whitegrid')
sns.set_context('paper', font_scale=1.8)
plt.rcParams['font.family'] = 'Helvetica'

palette = sns.color_palette().as_hex()
```

```python
data_folder = Path(os.path.expanduser('~')) / 'Documents' / 'PG_modelling'
assert data_folder.is_dir()
```

## AF3

```python
af3_results = pd.concat(
    [
        pd.read_csv(data_folder / 'Human_Lysozyme_C_P61626' / 'modelling' / 'af3' / 'scores.csv'),
        pd.read_csv(data_folder / 'ZoocinA_O54308' / 'modelling' / 'af3' / 'scores.csv'),
        pd.read_csv(data_folder / 'Lysostaphin_P10547' / 'modelling' / 'af3' / 'scores.csv'),
    ], 
    ignore_index=True,
)
len(af3_results)
```

```python
df = af3_results
ax = df.plot.scatter(
    x='confidence', 
    y='posebusters_score',
);
ax.set_title('AF3 confidence vs PoseBusters score')
ax.set_ylim((df['posebusters_score'].min() - 0.3, df['posebusters_score'].max() + 0.3))
ax.set_yticks(sorted(df['posebusters_score'].unique()));
```

```python
ax = af3_results[af3_results['posebusters_score'] == 20].plot.scatter(
    x='confidence', 
    y='aev_plig_pK',
);
ax.set_title('AF3 confidence vs pK score');
```
