```python
import os
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cwd = os.getcwd()
if cwd.endswith('notebook'):
    os.chdir('..')
    cwd = os.getcwd()
```

```python
sns.set_theme(palette='colorblind', font_scale=1.3)
palette = sns.color_palette().as_hex()

data_folder = Path('./data/')
assert data_folder.is_dir()
```

```python
folder = data_folder / 'pgb1_logistic_regression'

train_loss = np.load(folder / 'train_loss.npy', allow_pickle=True)
val_loss = np.load(folder / 'val_loss.npy', allow_pickle=True)

train_accuracy = np.load(folder / 'train_accuracy.npy', allow_pickle=True)
val_accuracy = np.load(folder / 'val_accuracy.npy', allow_pickle=True)
```

```python
f, axes = plt.subplots(2, 1, figsize=(12, 10))

ax1, ax2 = axes.flatten()

epochs = list(range(1, len(train_loss) + 1))

ax1.plot(epochs, train_loss, '-', color=palette[0])
ax1.plot(epochs, val_loss, '-', color=palette[1])
ax1.set_ylabel('Loss');

ax2.plot(epochs, train_accuracy, '-', color=palette[0])
ax2.plot(epochs, val_accuracy, '-', color=palette[1])
ax2.set_xlabel('epochs')
ax2.set_ylabel('Accuracy (%)');
```

```python
f.savefig(folder / 'metrics.png', dpi=300);
```
